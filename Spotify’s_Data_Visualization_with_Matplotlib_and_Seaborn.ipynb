```python
import matplotlib.pyplot as plt
```

```python
# creating first plot
# Data: Number of songs listened to each day
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
songs = [50, 70, 30, 85, 40]
# Create the plot
plt.plot(days, songs)
# Add a title and labels
plt.title('Songs Listened to Each Day')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Songs')
# Display the plot
plt.show()
```

```python
# creating Basic Plots
import seaborn as sns
# Data: Music genres and their popularity scores
genres = ['Pop', 'Rock', 'Jazz', 'Hip Hop', 'Classical']
popularity = [90, 75, 50, 85, 60]
# Create a bar chart
sns.barplot(x=genres, y=popularity)
# Add a title
plt.title('Popularity of Music Genres')
# Display the plot
plt.show()
```

```python
#Intermediate Plots: Scatter Plots, Histograms

# Data: Song tempo and danceability
tempo = [120, 130, 110, 115, 125]
danceability = [80, 70, 60, 90, 85]
# Create a scatter plot
plt.scatter(tempo, danceability)
# Add a title and labels
plt.title('Tempo vs Danceability')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Danceability Score')
# Display the plot
plt.show()
```

```python
#Advanced Plots: Heatmaps, Pair Plots, and Beyond
import numpy as np
# Data: Correlation matrix of different features
data = np.random.rand(10, 12)
sns.heatmap(data, annot=True)
# Add a title
plt.title('Correlation Heatmap')
# Display the plot
plt.show()
```

```python
#Customizing Your Visualizations
# Apply a Seaborn theme
sns.set_theme(style="darkgrid")
# Create a customized line plot
plt.plot(days, songs, color='purple', linestyle='--', marker='o')
# Customize labels and title
plt.title('Customized Song Listening Habits')
plt.xlabel('Days')
plt.ylabel('Number of Songs')
# Display the plot
plt.show()
```

```python
#Adding Annotations and Legends
# Adding annotations
plt.plot(days, songs)
plt.annotate('Peak Listening Day', xy=('Thursday', 85),
xytext=('Friday', 90),

arrowprops=dict(facecolor='black', shrink=0.05))

# Adding a legend
plt.legend(['Number of Songs'])
# Display the plot
plt.show()
```

```python
# Advanced Customization Techniques
# Creating subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
# First subplot
ax[0].plot(days, songs, color='blue')
ax[0].set_title('Songs per Day')
# Second subplot
ax[1].bar(days, songs, color='orange')
ax[1].set_title('Bar Chart of Songs per Day')
# Display the plots
plt.show()
```
